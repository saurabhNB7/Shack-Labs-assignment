```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
amazon=pd.read_csv('amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
amazon.head()
```

```python
flipkart=pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
flipkart.head()
```

```python
# dropping columns that are not necessary for comapring prices
amazon=amazon.drop(['uniq_id','crawl_timestamp','product_url','product_category_tree','pid','image','is_FK_Advantage_product','description','product_rating','overall_rating','brand','product_specifications'],axis=1)
```

```python
amazon.head()
```

```python
# dropping columns that are not necessary for comapring prices
flipkart=flipkart.drop(['uniq_id','crawl_timestamp','product_url','product_category_tree','pid','image','is_FK_Advantage_product','description','product_rating','overall_rating','brand','product_specifications'],axis=1)
```

```python
flipkart.head()
```

```python
flipkart.rename(columns={'product_name':'Product name in Flipkart'},inplace=True)
```

```python
flipkart.head()
```

```python
flipkart.rename(columns={'retail_price':'Retail Price in Flipkart'},inplace=True)
flipkart.rename(columns={'discounted_price':'Discounted Price in Flipkart'},inplace=True)
```

```python
flipkart.head()
```

```python
amazon.rename(columns={'product_name':'Retail Price in Amazon'},inplace=True)
amazon.rename(columns={'retail_price':'Discounted Price in Amazon'},inplace=True)
amazon.rename(columns={'discounted_price':'Discounted Price in Amazon'},inplace=True)
```

```python
amazon.head()
```

```python
# Concatenating amazon and flipkart table for price comparison side by side
new=pd.concat([flipkart, amazon], axis=1)
```

```python
new.head(20)
```

```python
# final result in single table format to new csv file named as assign_2_solution
new.to_csv('Part 2: P.csv')
```
